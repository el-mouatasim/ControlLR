# Plotter

```python
import matplotlib.pyplot as plt
import re
import numpy as np
import csv
```

```python
#fig = plt.figure(figsize=(10,5))

def read(name):
    f = open(name, 'r')
    file = f.read()
    file = re.sub('\\[', '', file)
    file = re.sub('\\]', '', file)
    f.close()

    return [float(i) for idx, i in enumerate(file.split(','))]


def draw(modn,mode):
        
    if mode == 'train loss':
        clr = read('./results/{}/train_loss_ControlLR.txt'.format(modn))
        adam = read('./results/{}/train_loss_StepLR.txt'.format(modn))
        ExponentialLR = read('./results/{}/train_loss_ExponentialLR.txt'.format(modn))
        plt.plot(clr, 'r', label='ControlLR')
        plt.plot(adam, 'b', label='StepLR')
        plt.plot(ExponentialLR, 'g', label='ExponentialLR')
        plt.legend(loc='lower left')

    elif mode == 'valid loss':
        clr = read('./results/{}/valid_loss_ControlLR.txt'.format(modn))
        adam = read('./results/{}/valid_loss_StepLR.txt'.format(modn))
        ExponentialLR = read('./results/{}/valid_loss_ExponentialLR.txt'.format(modn))
        plt.plot(clr, 'r', label='ControlLR')
        plt.plot(adam, 'b', label='StepLR')
        plt.plot(ExponentialLR, 'g', label='ExponentialLR')
        plt.legend(loc='lower left')
  

    elif mode == 'valid acc':
        clr = read('./results/{}/valid_acc_ControlLR.txt'.format(modn))
        adam = read('./results/{}/valid_acc_StepLR.txt'.format(modn))
        ExponentialLR = read('./results/{}/valid_acc_ExponentialLR.txt'.format(modn))
        plt.plot(clr, 'r', label='ControlLR')
        plt.plot(adam, 'b', label='StepLR')
        plt.plot(ExponentialLR, 'g', label='ExponentialLR')
        plt.legend(loc='best')
        
    plt.xlabel('epoch')
    plt.ylabel(mode)
    plt.title('training result of Adam in model {}'.format(modn))
    plt.grid(True, which='both', axis='both')
    plt.savefig('./png/{}/{}.png'.format(modn, mode))
    plt.show()
modn = 'Clip'
draw(modn, mode='train loss')
draw(modn, mode='valid loss')
    
draw(modn, mode='valid acc')    
```
